```python
# Importing of libaries
import pandas as pd
import seaborn as sb
```

```python
# Creating a DataFrame named df and loading our csv file which contains the dataset(TWO_CENTURIES_OF_UM_RACES.csv) 
df = pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv')
```

```python
# Looking at the first 10 Rows in the dataframe(0-9)
df.head(10)
```

```python
""" Filtering Data"""
# Looking for USA races & athlete club(s) for athlete(s) that covered a distance of 50km or 50mi in 2020
# Creating a variable named filt to filter out the requested data
filt = (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA') & (df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020)
```

```python
#Using the pandas loc function to locate the rows of the filter(filt) as well as showing the columns 
#we want to be associated with the rows.
df.loc[filt,['Event name','Athlete club','Event distance/length','Year of event']]
```

Creating New Dataframe from Pre-existing Dataframe and Cleaning up the Data

```python
# Creating a new dataframe that only stores races ran in the USA
usa_races = df[(df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA') & (df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020)]
```

```python
# Looking at the number of rows and columns in the usa_races dataframe (26,090 Rows) X (13 Columns)
usa_races.shape
```

```python
# Looking at the first 5 rows of the usa_races dataframe
usa_races.head(5)
```

```python
# Removing (USA) from the event name column in the usa_races dataframe as it seems redundant
usa_races['Event name'] = usa_races['Event name'].str.split('(').str.get(0)
```

```python
usa_races.head()
```

```python
# Creating a new column for athlete age 
usa_races['Age'] = usa_races['Year of event'] - usa_races['Athlete year of birth']
```

```python
# Removing the 'h' from athletes performance 
usa_races['Athlete performance'] = usa_races['Athlete performance'].str.split(' ').str.get(0)
```

```python
usa_races.dtypes
```

```python
usa_races.head()
```

```python
usa_races.columns
```

```python
# Data Cleaning
# drop columns: # Athlete club, Athlete Country, Athlete Age Category

usa_races = usa_races.drop(columns=['Athlete club','Athlete country','Athlete age category'])
```

```python
usa_races
```

```python
# Looking for columns that contains null values 
usa_races.isna().sum()
```

```python
#Dropping/Deleting all rows that has null values
usa_races = usa_races.dropna()
```

```python
usa_races
```

```python
# Looking for duplicate Values(No duplicates found in our dataframe)
usa_races[usa_races.duplicated() == True]
```

```python
# Reset Index
usa_races.reset_index(drop= True,inplace=True)
```

```python
# Renaming column: Age ------------> Athlete age
usa_races.rename(columns={'Age':'Athlete age'},inplace=True)
```

```python
# Changing Data type
usa_races['Athlete age'] = usa_races['Athlete age'].astype(int)
```

```python
#Dropping 'Athlete year of birth' column
usa_races = usa_races.drop(columns=['Athlete year of birth'])
```

```python
# Changing Data type
usa_races['Athlete average speed'] =  usa_races['Athlete average speed'].astype(float)
```

```python
usa_races
```

```python
usa_races.columns
```

```python
# Renaming columns
usa_races.rename(columns={'Year of event':'year','Event dates':'date','Event distance/length':'distance_covered',
                          'Event number of finishers':'number_of_finishers','Athlete gender':'gender',
                          'Athlete average speed':'average_speed','Athlete ID':'athlete_id','Athlete age':'age',
                          'Athlete performance':'athlete_performance','Event name':'event_name'},inplace=True)
```

```python
# Changed column names
usa_races.columns
```

```python
# Reordering the data column
usa_races = usa_races[['date','athlete_id','gender','age','event_name','distance_covered','average_speed','athlete_performance','number_of_finishers','year']]
```

```python
# Dropping 'year' column 
usa_races.drop(columns='year',inplace=True)
```

```python
# Renaming  'distance_covered' column to 'distance'
usa_races.rename(columns={'distance_covered':'distance'},inplace= True)
```

```python
# what's the difference in speed for the 50km,50mi male to female
usa_races.groupby(['distance','gender'])['average_speed'].mean()
```

```python
# What age group are the best in the 50m Distance (20 + Distance min) (show 10)
usa_races.query('distance== "50mi"').groupby('age')['average_speed'].agg(['mean','count']).sort_values('mean',ascending=False).query('count>19').head(10)
```

```python
# What age group are the worst in the 50m Distance (20 + Distance min) (show 15) 
usa_races.query('distance== "50mi"').groupby('age')['average_speed'].agg(['mean','count']).sort_values('mean',ascending=True).query('count>19').head(15)
```

```python
# Which season is easier to run a race 
usa_races['race_month'] = usa_races['date'].str.split('.').str.get(1).astype(int)
usa_races['race_season'] = usa_races['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter' )
usa_races.groupby('race_season')['average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```

```python
# 50mi only
usa_races.query('distance == "50mi"').groupby('race_season')['average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```
